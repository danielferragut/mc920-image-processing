## Intensisty transformation
### Negative of a image
Given a image, for every pixel we want its complement in relation to 255, e.g for a pixel with value 1, swap its value for 255 - 1, 254.

```python
# img still holds the city png
plt.axis('off')
imgplot = plt.imshow(255 - img, cmap='gray')
```

### Pixel intensity between [100,200]
Given a image, every pixel value needs to be in the [100, 200] range, e.g 0 -> 100,  255 -> 200. So for a pixel with value x, it will be transformed to a value y such that :
\begin{equation}\label{eq:}
x = \frac{100}{255}y + 100
\end{equation}


```python
new_img = (100/255*img) + 100
# Two subplots, unpack the axes array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.imshow(img, cmap='gray')
ax1.set_title('Normal image')
ax2.imshow(new_img, cmap='gray')
ax2.set_title('Transformed image')
plt.show()
```

### Invert even rows
Given a image, invert every pixel in every even row such that the starting pixel is swapped by the last one and vice-versa.

```python
inverted_image = img.copy()
# Arrays start at 0, so even rows are actually the odd ones
inverted_image[1::2, :] = img[1::2, ::-1]
# Two subplots, unpack the axes array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.imshow(img, cmap='gray')
ax1.set_title('Normal image')
ax2.imshow(inverted_image, cmap='gray')
ax2.set_title('Inverted image')
plt.show()
```

### Reflect half the image
Given a image, the lower half has to be a reflection of the upper half.

```python
img_half = img.shape[0] // 2

reflected_image = img.copy()
reflected_image[img_half:] = img[img_half-1::-1, :]

# Two subplots, unpack the axes array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.imshow(img, cmap='gray')
ax1.set_title('Normal image')
ax2.imshow(reflected_image, cmap='gray')
ax2.set_title('Reflected image')
plt.show()
```
